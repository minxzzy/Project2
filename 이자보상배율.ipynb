```python
df = df[['Symbol', 'Name', '회계년','영업이익(천원)','이자비용(비영업)(천원)']]
df
```

```python
# 이자보상비용 계산
df['이자보상배율'] = df['영업이익(천원)'] / df['이자비용(비영업)(천원)']

# 결측치제거
df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
df.to_csv('csv/이자보상배율.csv', encoding='utf-8-sig')
```
