```python
df = pd.read_csv('csv/feature/2차_feature(1446).csv', index_col=0)
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
winsorizing(train_df)
```

```python
from sklearn.preprocessing import StandardScaler

X = train_df.iloc[:, 6:]
y = train_df['label_shifted']

scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
# X_test_standardized = scaler.transform(X_test)
scaled_X_df = pd.DataFrame(scaled_X, columns=X.columns)
```

## 전진선택법

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# 데이터 로드
# X_train = train_df.iloc[:,6:]
# y_train = train_df['label_shifted']  

# # Test 데이터에서 Features(X)와 Target(y) 나누기
# X_test = test_df.iloc[:,6:]
# y_test = test_df['label_shifted'] 

X = scaled_X_df.iloc[:, 6:]
y = train_df['label_shifted']

# 모델 정의
model = LogisticRegression(max_iter=200)

# RFE 객체 생성 (선택할 특징의 개수 지정)
rfe = RFE(estimator=model, n_features_to_select=30)

# RFE 학습
rfe.fit(X, y)

# 선택된 특징 출력
print("선택된 특징 (Ranking):", rfe.ranking_)
# print("선택된 특징 (Support):", rfe.support_)
print("선택된 특징 이름:", X.columns[rfe.support_])
```

## 후진제거법

```python
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold

# RFECV 객체 생성 (교차 검증 전략 지정)
rfecv = RFECV(estimator=model, step=1, cv=StratifiedKFold(5), scoring='f1')

X = scaled_X_df.iloc[:, 6:]
y = train_df['label_shifted']

# RFECV 학습
rfecv.fit(X, y)

# 선택된 특징 출력
print("선택된 특징 (Ranking):", rfecv.ranking_)
# print("선택된 특징 (Support):", rfecv.support_)
print("선택된 특징 이름:", X.columns[rfecv.support_])
print("최적의 특징 개수:", rfecv.n_features_)
```

## Ridge, Lasso

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import SelectFromModel

# 데이터 로드
# X_train = train_df.iloc[:,6:]
# y_train = train_df['label_shifted']  

# X_test = test_df.iloc[:,6:]
# y_test = test_df['label_shifted'] 

X = scaled_X_df.iloc[:, 6:]
y = train_df['label_shifted']

# Lasso 회귀를 사용한 특징 선택
lasso = Lasso(alpha=0.01, random_state=42)
lasso.fit(X, y)

y_pred = lasso.predict(X)
mse = mean_squared_error(y, y_pred)

print("Lasso 회귀의 MSE:", mse)
selected_features_lasso = X.columns[lasso.coef_ != 0]
print("Lasso를 사용한 선택된 특징:", selected_features_lasso)

print('\n================================================================================================\n')

# Ridge 회귀를 사용한 특징 선택
ridge = Ridge(alpha=1.0, random_state=42)
ridge.fit(X, y)
model = SelectFromModel(ridge, prefit=True)
selected_features_ridge = X.columns[model.get_support()]
print("Ridge를 사용한 선택된 특징:", selected_features_ridge)

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 상관 행렬 계산
corr = df.corr()

# 히트맵 그리기
# plt.figure(figsize=(20, 10))
# sns.heatmap(corr, annot=True, cmap="YlGnBu", cbar=True)

# plt.title("Heatmap of Financial Ratios")
# plt.show()

mask = np.triu(np.ones_like(corr, dtype=bool))

# 히트맵 그리기
plt.figure(figsize=(30, 20))
sns.heatmap(corr, annot=True, cmap="YlGnBu", cbar=True, mask=mask, center=0, vmin=-1, vmax=1)

# plt.title("Heatmap of Financial Ratios")
plt.show()
```

```python
import pandas as pd
import numpy as np


# 상관 행렬 계산
corr = df.corr()

# 대각선을 제외한 상관 계수 값들을 1차원 배열로 변환
corr_values = corr.where(~np.eye(corr.shape[0],dtype=bool)).stack()

# 가장 큰 5개 값과 가장 작은 5개 값 추출
largest_5 = corr_values.nlargest(10)
smallest_5 = corr_values.nsmallest(10)

print("가장 큰 5개 상관 계수:")
for index, value in largest_5.items():
    print(f"{index}: {value:.4f}")

print("\n가장 작은 5개 상관 계수:")
for index, value in smallest_5.items():
    print(f"{index}: {value:.4f}")
```
