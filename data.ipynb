```python
import pandas as pd
```

```python
kospi = pd.read_csv('csv/코스피.csv')
kosdaq = pd.read_csv('csv/코스닥.csv')
list = pd.read_csv('csv/기업목록.csv', index_col=0)
```

```python
list = list[~list['Symbol'].isin(['A000570',
'A002950',
'A000830',
'A104110',
'A057110',
'A021570',
'A131390',
'A002300',
'A032380',
'A012830'])]
list
```

```python
kospi['시장'] = 'KOSPI'
kosdaq['시장'] = 'KOSDAQ'
```

```python
kosdaq
```

```python
data = pd.concat([kospi, kosdaq])
data.reset_index(drop=True, inplace=True)
```

```python
data = pd.merge(list, data, on=['Symbol', '회계년'], how='left')
data
```

```python
duplicated_rows = data[data.duplicated(subset=['Name', '회계년'], keep=False)]
print(duplicated_rows)
```

```python
duplicated_rows.to_csv('중복데이터.csv', encoding='utf-8-sig')
```
