### 차입부채

```python
datalist = full[['Symbol','Name','회계년','이자비용(천원)','영업수익(천원)','영업이익(천원)','현금흐름(천원)','*총차입부채(천원)','EBITDA(천원)']]
# mean = datalist.groupby(['Name','Symbol','회계년'])['*총차입부채(천원)'].mean().reset_index()
datalist['*총차입부채(천원)_전년도'] = datalist.groupby('Symbol')['*총차입부채(천원)'].shift(1)
datalist
```

```python
datalist['총차입부채(천원)_평균'] = (datalist['*총차입부채(천원)'] + datalist['*총차입부채(천원)_전년도']) / 2
datalist
```

```python
data = pd.merge(datalist, list, on=['Name','Symbol','회계년'])
data
```

```python
data['차입부채조달금리(평균)(%)'] = (data['이자비용(천원)'] / data['총차입부채(천원)_평균']) * 100
data['차입부채조달금리(기말)(%)'] = (data['이자비용(천원)'] / data['*총차입부채(천원)']) * 100

data['차입부채/영업수익(%)'] = (data['*총차입부채(천원)'] / data['영업수익(천원)']) * 100
data['차입부채/영업이익(배)'] = (data['*총차입부채(천원)'] / data['영업이익(천원)'])
data['차입부채/EBITDA(배)'] = (data['*총차입부채(천원)'] / data['EBITDA(천원)'])

data['현금흐름/총차입부채(%)'] = (data['현금흐름(천원)'] / data['총차입부채(천원)_평균']) * 100
data
```

```python
data.to_csv('csv/재무/연결_차입부채.csv',encoding='utf-8-sig')
```

### 유동, 고정비율 등

```python
datalist = weak_data[[
    'Symbol', 'Name', '회계년', '유동자산(천원)', '당좌자산(천원)', '유동부채(천원)', 
    '비유동자산(천원)', '자본금(천원)', '자본잉여금(천원)', '주식발행초과금(천원)', 
    '기타포괄손익누계액(천원)', '이익잉여금(천원)', '사채(천원)', '장기차입금(천원)', 
    '장기매입채무및기타채무(천원)', '기타장기충당부채(천원)', '이연법인세부채(천원)', 
    '매출채권(천원)', '재고자산(천원)', '매입채무(천원)', '총자본(천원)'
]]

datalist['자기자본(천원)'] = (
    datalist['자본금(천원)'] + datalist['자본잉여금(천원)'] + datalist['주식발행초과금(천원)'] + 
    datalist['기타포괄손익누계액(천원)'] + datalist['이익잉여금(천원)']
)

datalist['비유동부채(천원)'] = (
    datalist['사채(천원)'] + datalist['장기차입금(천원)'] + datalist['장기매입채무및기타채무(천원)'] + 
    datalist['기타장기충당부채(천원)'] + datalist['이연법인세부채(천원)']
)

datalist['유동비율'] = (datalist['유동자산(천원)'] / datalist['유동부채(천원)']) * 100
datalist['비유동비율'] = (datalist['비유동자산(천원)'] / datalist['자기자본(천원)']) * 100
datalist['당좌비율'] = (datalist['당좌자산(천원)'] / datalist['유동부채(천원)']) * 100
datalist['유동부채비율'] = (datalist['유동부채(천원)'] / datalist['자기자본(천원)']) * 100
datalist['비유동부채비율'] = (datalist['비유동부채(천원)'] / datalist['자기자본(천원)']) * 100
datalist['매출채권대 재고자산비율'] = (datalist['매출채권(천원)'] / datalist['재고자산(천원)']) * 100
datalist['매입채무대 재고자산비율'] = (datalist['매입채무(천원)'] / datalist['재고자산(천원)']) * 100
datalist['순운전자본대 총자본비율'] = ((datalist['유동자산(천원)'] - datalist['유동부채(천원)']) / datalist['총자본(천원)']) * 100

datalist
```

### 유보율, ROE

```python
datalist = weak_data[['Symbol','Name','회계년','*유보액(천원)','자본금(천원)','총자산(천원)']]
datalist['유보율(%)'] = weak_data['*유보액(천원)'] / weak_data['자본금(천원)'] * 100
datalist['유보액/총자산(%)'] = weak_data['*유보액(천원)'] / weak_data['총자산(천원)'] * 100

datalist['ROE(영업이익)'] = weak_data['영업이익(연율화)(천원)'] / weak_data['총자본(평균)(천원)'] * 100
datalist['ROE(세전계속사업이익)'] = weak_data['세전계속사업이익(연율화)(천원)'] / weak_data['총자본(평균)(천원)'] * 100
datalist['ROE(당기순이익)'] = weak_data['당기순이익(연율화)(천원)'] / weak_data['총자본(평균)(천원)'] * 100
datalist['ROE(지배주주순이익)'] = weak_data['지배주주순이익(연율화)(천원)'] / weak_data['지배주주지분(평균)(천원)'] * 100
datalist['ROE(지배주주순이익)기말'] = weak_data['지배주주순이익(연율화)(천원)'] / weak_data['지배주주지분(천원)'] * 100

datalist
```

```python
datalist.to_csv('csv/재무/연결_유보율_ROE.csv', encoding='utf-8-sig')
```
