```python
import pandas as pd
df = pd.read_csv('csv/상장일자.csv')
df['최초상장년도'] = df['최초상장일자'].astype(str).str[:4]
df['상장년도'] = df['상장일자'].astype(str).str[:4]
df
```

```python
# 최초상장년도와 상장년도가 같지 않은 행만 추출
df_diff = df[df['최초상장년도'] != df['상장년도']]
df_diff
```

```python
pre = pd.read_csv('csv/240522상장해.csv',encoding='cp949')
pre['코넥스'] = pre['코넥스'].fillna(0)
pre
```

```python
# 코넥스 상장X (0) 상장해 = 최초상장년도, 코넥스에 상장O (1) 상장해 = 상장년도
pre['상장해'] = ''
pre['상장해'] = pre.apply(lambda row: row['최초상장년도'] if row['코넥스'] == 0 else row['상장년도'], axis=1)
pre.drop(columns='코넥스', inplace = True)
pre
```

```python
# 상장일자.csv, 240522상장해.csv
pre['최초상장년도'] = pre['최초상장일자'].astype(str).str[:4]
pre['상장년도'] = pre['상장일자'].astype(str).str[:4]

df = pd.merge(df, pre, on = ['Symbol', 'Name', '최초상장일자', '상장일자', '최초상장년도', '상장년도'], how = 'left')
df
```

```python
# 결측치 19개
df.loc[df['최초상장일자'].isna(),:]
```

```python
def set_listing_year(row):
    if pd.isna(row['최초상장일자']):
        return 2025
    else:
        return row['상장해']

df['상장해'] = df.apply(set_listing_year, axis=1)
df
```

```python
# 상장해가 결측치인 경우 최초상장년도 값으로 채우기
df['상장해'] = df.apply(lambda row: row['최초상장년도'] if pd.isna(row['상장해']) else row['상장해'], axis=1)
df
```
