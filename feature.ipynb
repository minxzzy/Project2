```python
import pandas as pd
feature = pd.read_csv('csv/코스피코스닥연결재무비율.csv', index_col=0)
feature.reset_index(inplace=True)
label = pd.read_csv('csv/최종_label.csv', index_col=0)
label.reset_index(drop=True, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
```

```python
label
```

```python
columns_to_drop = [col for col in feature.columns if '분기' in col or '3년' in col or '5년' in col]
feature = feature.drop(columns=columns_to_drop)
feature
```

```python
feature = feature.drop(columns=['부가가치(천원)', '자본생산성(%)', '노동생산성(천원)'])
feature
```

```python
label
```

```python
df = pd.merge(label, feature, on=['Symbol', 'Name', '회계년'], how='left')
df
```

```python
df.to_csv('csv/feature/후보피처1.csv',encoding='utf-8-sig')
```
